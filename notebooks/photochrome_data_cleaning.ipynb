```python
from bs4 import BeautifulSoup as BS
import json
import glob
import pandas as pd
import csv
import os
from tqdm import tqdm
```

```python
#make list of json files
all_json = glob.glob("*.json")
len(all_json)
```

```python
#creat csv from JSON files

filename = 'photochrommetadata.csv'
no_source = 'no source'
locationdef = 'na'
    
#open csv with the rows source, author, title, illustration
with open(filename, 'w') as csvfile: 
    csvwriter = csv.writer(csvfile, delimiter='\t')
    csvwriter.writerow(["title","contributor","date","lot","country","city","locationdef","source","basename","ur"])
    

    for file_ in tqdm(all_json):
        basename = os.path.basename(file_)
        basename = basename[:-4] + 'jpg'        
        with open(file_, 'r') as json_file:
            data = json.load(json_file)
            try:
                name = data['item']['contributor_names']
            except:
                name = None
                name = 'no contributor'
            try:
                Lot = data['item']['call_number']
                Lot = Lot.split(',', 1)[0]  # maxsplit = 1; 
            except:
                Lot = None
                Lot = 'no lot'
            try:
                country = data['item']['location_country']
                country = str(country)
            except:
                country = None
                country = 'no country'
            try:
                city = data['item']['location_city']
                city = str(city)
            except:
                city = None
                city = 'no city'
            try:
                source = data['item']['source_collection']
            except:
                soure = None
                source = 'no source'
            try:
                title = data['item']['title']
                title = str(title)
            except:
                title = 'no title in json'
            try:
                date = data['item']['date']
            except:
                date = 'no date in json'
            try:
                url = data['item']['id']
            except:
                url = 'no url in json'
        csvwriter.writerow([title,name,date,Lot,country,city,locationdef,source,basename,url])

csvfile.close()
```

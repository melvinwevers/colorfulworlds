```python
# drop non-autochrome based on 'Procédé technique'
values = ['Négatif sur verre noir et blanc', 'Filmcolor', 'Positif noir et blanc', 'Positif sur verre noir et blanc', 'Positif sur verre noir et blanc viré', 'Négatif noir et blanc', 'Omnicolore', 'Autochrome stéréo', 'Procédé à réseau indéterminé']
df = df[df['Procédé technique'].isin(values) == False]
```

```python
# Tried translating with GoogleTrans but did not work, use dictionary instead 
transC = {"Japon" : 'Japan', "Belgique" : 'Belgium', "Grèce" : 'Greece', "Turquie" : 'Turkey', "Suisse" : 'Switzerland', "Allemagne" : 'Germany', "Algérie": 'Algeria', "Italie" : 'Italy', "Inde": 'India', "Bénin" : 'Benin', "Maroc" : 'Morocco', "Grande-Bretagne" : 'UK', "Egypte" : 'Egypt', "Chine" : 'China', "Irak" : 'Iraq', "Espagne" : 'Spain', "Syrie" : 'Syria', "Liban" : 'Lebanon', "Tunisie" : 'Tunesia', "Israël" : 'Israel', "Bosnie-Herzégovine" : 'Bosnia and Herzegovina', "Pays-Bas" : 'Netherlands', "Norvège" : 'Norway', "Macédoine (FYROM)" : 'Macedonia', "Suède" : 'Sweden', "Mongolie" : "Mongolia", "Jordanie" : 'Jordan', "Albanie" : 'Albania', "Arabie Saoudite" : 'Saudi Arabia', "Serbie" : 'Serbia', "Brésil" : 'Brazil', "Russie" : 'Russia', "Chypre" : "Cyprus", "Monténégro" : 'Montenegro', "Bulgarie" : 'Bulgaria', "Croatie" : 'Croatia', "Hongrie" : 'Hungary', "Autriche" : "Austria", "Tchéquie" : 'Czech Republic', "Monaco [Principauté]" : 'Monaco'}
TransR = {"Proche-Orient" : 'Middle East', "Amérique du Sud" : 'South America'}
df=df.replace({"Pays": transC})
df=df.replace({"Région": TransR})
```

```python
# Replace months with english months
df['Date de prise de vue'] = df['Date de prise de vue'].str.replace('janvier', 'january')
df['Date de prise de vue'] = df['Date de prise de vue'].str.replace('février', 'February')
df['Date de prise de vue'] = df['Date de prise de vue'].str.replace('mars', 'March')
df['Date de prise de vue'] = df['Date de prise de vue'].str.replace('avril', 'April')
df['Date de prise de vue'] = df['Date de prise de vue'].str.replace('mai', 'mai')
df['Date de prise de vue'] = df['Date de prise de vue'].str.replace('juin', 'June')
df['Date de prise de vue'] = df['Date de prise de vue'].str.replace('juillet','July')
df['Date de prise de vue'] = df['Date de prise de vue'].str.replace('aout', 'August')
df['Date de prise de vue'] = df['Date de prise de vue'].str.replace('septembre', 'September')
df['Date de prise de vue'] = df['Date de prise de vue'].str.replace('octobre', 'October')
df['Date de prise de vue'] = df['Date de prise de vue'].str.replace('novembre', 'November')
df['Date de prise de vue'] = df['Date de prise de vue'].str.replace('décembre', 'December')
```

```python
# copry relevant colums 
df1 = df[['Légende', 'Date de prise de vue', 'Pays', 'filename', 'Photo']].copy()
```

```python
#Rename columns for clarity
df = df1.rename(columns={"Légende": "title", "Date de prise de vue": "date", "Pays" : "location", "Photo" : "url" })
```

```python
#save df to 
df.to_csv('Autochromedata.csv', sep='\t')
```
